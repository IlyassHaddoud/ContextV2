```python
import os
import tensorflow
import numpy as np
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Convolution2D, MaxPool2D, ZeroPadding2D,Flatten,Dense,Dropout
from keras import optimizers
from keras.preprocessing import image
import matplotlib.pyplot as plt
from PIL import Image

```

```python
# Répertoires des données d'entraînement et de validation

train_data_dir = 'drive/MyDrive/data/train'
validation_data_dir = 'drive/MyDrive/data/validation'
```

```python
# Définition de la largeur et de la hauteur des images d'entrée

img_width, img_height = 128, 128

# Création d'un générateur d'images avec prétraitement

datagen = ImageDataGenerator(rescale=1./255)

# Création d'un générateur d'images pour les données d'entraînement

train_generator = datagen.flow_from_directory(
        train_data_dir,
        target_size=(img_width,img_height),
        batch_size = 32,
        class_mode = 'categorical')
print(train_generator.class_indices)

# Création d'un générateur d'images pour les données de validation

validation_generator = datagen.flow_from_directory(
        validation_data_dir,
        target_size=(img_width,img_height),
        batch_size = 32,
        class_mode = 'categorical')
print(validation_generator.class_indices)
```

```python
# Création du modèle CNN

classifier= Sequential()

classifier.add(Convolution2D(32, kernel_size=(3, 3), strides=(1, 1), input_shape=(128,128,3), activation='relu'))
classifier.add(Dropout(0.25))

classifier.add(MaxPool2D(pool_size=(2,2)))

classifier.add(Convolution2D(64, kernel_size=(3, 3), strides=(1, 1), activation='relu'))
classifier.add(Dropout(0.25))

classifier.add(MaxPool2D(pool_size=(2,2)))

classifier.add(Flatten())

classifier.add(Dense(256, activation='relu'))
classifier.add(Dropout(0.5))

classifier.add(Dense(2, activation='softmax'))

learning_rate = 0.00001
optimizer = tensorflow.keras.optimizers.RMSprop(learning_rate=learning_rate)

```

```python
# Compiler le modèle

classifier.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=["accuracy"])
```

```python
# Entraînement du modèle

model_history = classifier.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=10,
    validation_data=validation_generator,
    validation_steps=len(validation_generator)
    )
```

```python
# Visualisation des résultats d'entraînement

accuracy = model_history.history['accuracy']
val_accuracy  = model_history.history['val_accuracy']

loss = model_history.history['loss']
val_loss = model_history.history['val_loss']
plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.plot(accuracy, label = "Training accuracy")
plt.plot(val_accuracy, label="Validation accuracy")
plt.legend()
plt.title("Training accuracy vs validation accuracy")

plt.subplot(2,2,2)
plt.plot(loss, label = "Training loss")
plt.plot(val_loss, label="Validation loss")
plt.legend()
plt.title("Training loss vs validation loss")

plt.show()
```

```python
# Prédiction sur une image

image_path = 'simo.jpeg'
img = Image.open(image_path)
img = img.resize((img_width, img_height))
img = np.array(img)
img = img.reshape(1, img_width, img_height, 3)
img = img / 255.0

prediction = classifier.predict(img)
class_index = np.argmax(prediction)
class_labels = train_generator.class_indices
class_label = [k for k, v in class_labels.items() if v == class_index][0]

plt.imshow(img[0])
plt.title(f"Prédiction : {class_label}")
plt.axis('off')
plt.show()
```
